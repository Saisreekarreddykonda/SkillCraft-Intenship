```python
# Task 04: UK Road Safety Accident Analysis
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/UK_Accident.csv", low_memory=False)

print("Dataset shape:", df.shape)
print(df.head())

# Preprocessing
# Convert date and time
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce').dt.hour
df['DayOfWeek'] = df['Date'].dt.day_name()
```

```python
# Accidents by Day of Week
dow_counts = df['Day_of_Week'].value_counts().sort_index()
plt.figure(figsize=(8,4))
dow_counts.plot(kind='bar', color='cornflowerblue')
plt.title("Accidents by Day of Week")
plt.xlabel("Day of Week (1=Sunday)")
plt.ylabel("Accident Count")
plt.show()
```

```python
# Accidents by Hour
hourly = df['Hour'].value_counts().sort_index()
plt.figure(figsize=(10,4))
hourly.plot(kind='bar', color='darkorange')
plt.title("Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Accident Count")
plt.show()
```

```python
# Weather Conditions
weather_counts = df['Weather_Conditions'].value_counts().head(10)
plt.figure(figsize=(10,5))
weather_counts.plot(kind='bar', color='seagreen')
plt.title("Top 10 Weather Conditions in Accidents")
plt.xlabel("Weather Condition")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Road Surface Conditions
road_surface = df['Road_Surface_Conditions'].value_counts()
plt.figure(figsize=(6,4))
road_surface.plot(kind='bar', color='firebrick')
plt.title("Accidents by Road Surface Condition")
plt.xlabel("Road Surface")
plt.ylabel("Count")
plt.show()
```

```python
# Accident Hotspots
plt.figure(figsize=(8,6))
plt.hexbin(df['Longitude'], df['Latitude'], gridsize=60, bins='log', cmap='Reds')
plt.colorbar(label='log(#accidents)')
plt.title("Accident Hotspots (Hexbin)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
